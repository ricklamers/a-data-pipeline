```python
import orchest
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm
```

```python
dataset = orchest.get_inputs()['dataset']
```

```python
data = dataset['data']; data.shape
```

```python
cm = matplotlib.cm.get_cmap()
```

```python
max_target = np.max(dataset['target'])

for i, e in enumerate(data):
    target = dataset['target'][i]
    plt.plot(e, color=cm(target/max_target))
```
